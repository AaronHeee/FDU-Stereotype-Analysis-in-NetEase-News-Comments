```python
province = set(df_freq.dist)
res = {}
for p in province:
    res[p] = sum(df_freq[(df_freq.src==p)&(df_freq.polar>0)&(df_freq.src==df_freq.dist)].freq)/sum(df_freq[(df_freq.src==p)&(df_freq.src==df_freq.dist)].freq)
from pyecharts import WordCloud
name = [key for key in res.keys()]
value = [res[key] for key in res.keys()]
wordcloud = WordCloud(width=800, height=620)
wordcloud.add("", name, value, word_size_range=[20, 100],shape='diamond')
wordcloud
```

```python
from pyecharts import Map

province = set(df_freq.dist)
res = {}
for p in province:
    res[p] = sum(df_freq[(df_freq.src==p)&(df_freq.polar<0)&(df_freq.src!=df_freq.dist)].freq)/sum(df_freq[(df_freq.src==p)&(df_freq.src!=df_freq.dist)].freq)
from pyecharts import WordCloud
attr = [key for key in res.keys()]
value = [100*res[key] for key in res.keys()]
map = Map("全国地域负面评价分布图", width=800, height=600)
map.add("", attr, value, maptype='china', is_label_show=True, is_visualmap=True, visual_range=[25,50])
map.render()
```
